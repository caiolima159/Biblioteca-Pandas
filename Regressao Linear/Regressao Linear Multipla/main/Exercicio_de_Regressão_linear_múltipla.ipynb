```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
# Importando os dados do csv
df = pd.read_csv('/content/C+¦pia de CarPrice_Assignment.csv')

# Comando para exibir todas as colunas do df
pd.set_option('display.max_columns', None)
df.head()
```

```python
# Transformando os dados do tipo texto em dados numéricos com o LabelEncoder

# Criando uma função para transformar todas as colunas texto em colunas numéricas
def Encoder(df, lista):
  le = LabelEncoder()
  for i in lista:
    le.fit(df[i])
    df[i] = le.transform(df[i])
  return df

# Executando a função criada
textos = ['CarName', 'fueltype', 'aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']
Encoder(df, textos)
df.head()
```

```python
# Observando se há grandes variações entre os valores dos dados
df.describe()
```

Como existem grandes diferenças entre os valores, recomenda-se fazer uma normalização (deixar os valores entre 0 e 1)

```python
# Criando um df_norma que é um df em que os dados variam de 0-1
escala = MinMaxScaler()
df_norma = pd.DataFrame(escala.fit_transform(df), index=df.index, columns=df.columns)
df_norma.describe()
```

É facil ver que os dados estão próximos uns dos outros, e isso não acontecia antes.

```python
# Colunas diferentes da coluna target
n_target = (df_norma[df_norma.columns[df_norma.columns != 'price']].columns)

# Transformando em texto e juntando as colunas com um +
n_target = '+'.join(n_target)
n_target
```

```python
# Criando a função para usar no modelo OLS que tem como target a variavel price (ou seja, ele é o nosso y)
funcao = 'price~'+n_target
funcao
```

```python
# Criando um sumário
modelo = smf.ols(formula=funcao, data=df_norma).fit()
print(modelo.summary())
```

```python
# As colunas que vão em x são as que colunas que o P value era menor que 0.1

var_x = 'CarName+fueltype+carbody+drivewheel+enginelocation+carwidth+carheight+curbweight+enginesize+stroke+horsepower+peakrpm'.split('+')
var_x
```

```python
# Definindo o x e o y para realizar o treino e predict

x = df_norma[var_x]
y = df_norma[['price']]
```

```python
# Separando as variaveis em treino e teste, foram usados 20% dos dados para teste e 80% para treino

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.2, random_state=42)
lr = LinearRegression()
lr.fit(x,y)
```

```python
# Medindo em quantos % nosso modelo ta acertando os valores.

r_2 = lr.score(x,y)
r_2
# O nosso modelo acerta aproximadamente 89,46% das variaveis
```

```python
# Calculando as metricas de regressão para os valores de treino

y_pred_treino = lr.predict(x_treino)
print('MAE:', metrics.mean_absolute_error(y_treino, y_pred_treino))
print('MSE:', metrics.mean_squared_error(y_treino, y_pred_treino))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_treino, y_pred_treino)))
```

```python
# Calculando as metricas de regressão para os valores de teste

y_pred_teste = lr.predict(x_teste)
print('MAE:', metrics.mean_absolute_error(y_teste, y_pred_teste))
print('MSE:', metrics.mean_squared_error(y_teste, y_pred_teste))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_teste, y_pred_teste)))
```

Como a diferença foi pouca entre os resultados da base teste e a base de treino. Então quer dizer que nosso modelo performou bem.
